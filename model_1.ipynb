```python
charact_dict = {'I': (6, 'blue'), 'C': (4, 'green'), 'F': (9, 'red')}
d_central = nx.degree_centrality(G)

dict_nodes = {}
for i in range(G.number_of_nodes()):
    dict_nodes[f'node_{i}'] = {'type': 'I', 'consump': charact_dict['I'][0], 'n_c_neighbors': 0, 'd_centrality': d_central[i], 'color': charact_dict['I'][1]}

```

```python
attrs = {i: dict_nodes[f'node_{i}'] for i in range(G.number_of_nodes())}
nx.set_node_attributes(G, attrs)
```

```python
# Prob of cooperation
beta_ = 0.4

def prob_c(beta, m):
    return 1-(1-beta)**m

# Prob of free-rider
gamma_ = 0.3
```

```python
def change_status(beta, m, gamma):
    p_c = prob_c(beta, m)

    if p_c > gamma:
        return 'C'
    
    elif p_c < gamma:
        return 'F'
    
    else:
        return 'I'
```

```python
def get_number_c_neighbors(G, id_n):
    n_c = 0
    for neighbor in G.edges(id_n):
        type_n = nx.get_node_attributes(G, 'type')[neighbor[1]]
        
        if type_n == 'C':
            n_c += 1

    return n_c
```
